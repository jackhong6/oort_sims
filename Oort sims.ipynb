```python
import galpy.potential as pot
import astropy.units as u
from scipy.optimize import brentq, minimize
```

```python
N = 2 # number of arms
alpha = 10 * u.deg  # pitch angle, p in the paper
H = 0.18 * u.kpc  # arbitrary, not specified in the paper
Rs = 0.38
sp = pot.SpiralArmsPotential(amp=1, N=N, alpha=alpha, H=H, Rs=Rs, phi_ref=0)
phi_ref = -minimize(lambda phi: sp.dens(1, 0, phi[0]), x0=[0]).x[0]
sp = pot.SpiralArmsPotential(amp=1, N=N, alpha=alpha, H=H, Rs=Rs, phi_ref=phi_ref)

lp = pot.LogarithmicHaloPotential(amp=1, normalize=1)

omega_corot = brentq(lambda s: pot.lindbladR([lp], s*u.km/u.s/u.kpc, m='corot')-1, a=25, b=30)
omega_ILR = brentq(lambda s: pot.lindbladR([lp], s*u.km/u.s/u.kpc, m=-2)-1, a=40, b=50)
omega_OLR = brentq(lambda s: pot.lindbladR([lp], s*u.km/u.s/u.kpc, m=2)-1, a=5, b=15)
omega_L3 = brentq(lambda s: pot.lindbladR([lp], s*u.km/u.s/u.kpc, m=3)-1, a=12, b=17)
omega_Ln3 = brentq(lambda s: pot.lindbladR([lp], s*u.km/u.s/u.kpc, m=-3)-1, a=32, b=45)
```

```python
def linear_model(x, m, b):
    return m * x + b

def exp_model(x, A, r, p, d):
    return A * np.exp(r*(x-p)) + d

def log_model(x, A, r, p, d):
    return A * np.log(r * (x-p)) + d
```

```python
popt, pcov = curve_fit(linear_model, coldA1_omegas, sign(coldA1_data)*log(np.abs(coldA1_data)), p0=[-.2, .8])
popt
```
